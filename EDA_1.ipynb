```python
import pandas as pd

data = pd.get_dummies(data, columns = ['embarked'], drop_first = True)  #since the column has three categories and since its binary, we do drop first = True
print(data)
```

#**8. Dimension reduction**

Prinicipal Component Analysis

```python
print(data)
print(data.info())
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

#removing null values from numerical columns
data['age'].fillna(data['age'].mean(), inplace = True)

#standardising
Features = ['age', 'fare', 'family']
x = StandardScaler().fit_transform(data[Features])

#Applying PCA
pca = PCA(n_components = 2)  #reduces the data to 2 principal components.
pca_result = pca.fit_transform(x)

#add pca results to the dataframe
data['pc1'] = pca_result[:,0]
data['pc2'] = pca_result[:,1]
print(data)

#Plotting
sns.scatterplot(x = data['pc1'], y = data['pc2'], hue = data['survived'])
plt.show()
```

#**9. Data Scaling**

Standardization and Normalization

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

stdsca = StandardScaler()
data[['age_scaled']] = stdsca.fit_transform(data[['age']])

mimasca = MinMaxScaler()
data[['age_normalized']] = mimasca.fit_transform(data[['age']])

print(data)
```

#**10. Initial Modelling and Hypothesis Testing**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x = data[['age', 'fare', 'family']]  #input features
y = data['survived']   #output

#splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

#Calling the model
model = LogisticRegression()
model.fit(x_train, y_train)

#predicting
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'{accuracy:.2f}')
```
